```python
admet_data=pd.read_excel("../化合物ADMET.xlsx")
```

```python
admet_data['MN'].unique()
```

```python
def handleLabel_mn(dataframe):
    global anomaly_list
    #print(dataframe['MN'])
    if dataframe['MN']  != 1:
        return 'good'
    else:
        return 'bad'
    
def handleLabel_her(dataframe):
    global anomaly_list
    #global label_list  #在函数内部使用全局变量并修改它
    if dataframe['hERG']  != 1:
        return 'good'
    else:
        return 'bad'
```

```python
admet_data['MN']=admet_data.apply(lambda r:handleLabel_mn(r),axis=1)#每一行来传
admet_data['hERG']=admet_data.apply(lambda r:handleLabel_her(r),axis=1)
```

```python
admet_data['MN'].unique()
```

```python
admet_data['hERG'].unique()
```

```python
def handleLabel_imn(dataframe):
    global anomaly_list
    #print(dataframe['MN'])
    if dataframe['MN']  == 'good':
        return 1
    else:
        return 0
    
def handleLabel_iher(dataframe):
    #global label_list  #在函数内部使用全局变量并修改它
    if dataframe['hERG']  == 'good':
        return 1
    else:
        return 0
```

```python
admet_data['MN']=admet_data.apply(lambda r:handleLabel_imn(r),axis=1)#每一行来传
admet_data['hERG']=admet_data.apply(lambda r:handleLabel_iher(r),axis=1)
```

```python
def count_good_label(dataframe):
    #global label_list  #在函数内部使用全局变量并修改它
    if dataframe[1]+dataframe[2]+dataframe[3]+dataframe[4]+dataframe[5] >= 3:
        return 1
    else:
        return 0
```

```python
admet_data['own_3_good'] = admet_data.apply(lambda r:count_good_label(r),axis=1)
```
