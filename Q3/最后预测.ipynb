```python
import pandas as pd
import numpy as np
import joblib

test_df = pd.read_csv("Q3-test.csv")
target_label = ['Caco-2', 'CYP3A4', 'hERG', 'HOB', 'MN','SMILES']
class_label = ['Caco-2', 'CYP3A4', 'hERG', 'HOB', 'MN']

X = test_df.drop(labels=target_label,axis=1)
```

```python
caco_predictor = joblib.load("caco_rfc.model")
mn_predictor =joblib.load("mn_lgbm.model")
herg_predictor = joblib.load("herg_lgbm.model")
hob_predictor =joblib.load("hob_lgbm.model")
cyp3a4_predictor = joblib.load("cyp3a4_xgbc.model")
```

```python
caco_pred = caco_predictor.predict(X)
mn_pred = mn_predictor.predict(X)
herg_pred = herg_predictor.predict(X)
hob_pred = hob_predictor.predict(X)
cyp3a4_pred = cyp3a4_predictor.predict(X)
```

```python
type(caco_pred)
```

```python
#a = np.vstack((caco_pred,mn_pred))
a = pd.concat([pd.DataFrame(caco_pred),pd.DataFrame(cyp3a4_pred)
               ,pd.DataFrame(herg_pred),pd.DataFrame(hob_pred),pd.DataFrame(mn_pred)],axis=1
              ,names=class_label)
a.columns = class_label
a.to_csv('测试结果最终.csv',index=False)
a.head()
```
